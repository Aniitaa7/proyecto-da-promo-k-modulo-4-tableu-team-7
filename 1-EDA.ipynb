```python
# Contar valores duplicados en todo el dataset
duplicados = df.duplicated().sum()
print(f"Valores duplicados en el dataset: {duplicados}")
```

```python
# Calculamos los valores faltantes por columna
missing_values = df.isnull().sum()

# Calculamos el porcentaje de valores faltantes
missing_percentage = (missing_values / len(df)) * 100
```

```python

# Creamos un DataFrame de valores faltantes
missing_values_df = pd.DataFrame({
    'Column': missing_values.index,
    'Missing Values': missing_values.values,
    'Percentage': missing_percentage
}).sort_values(by='Missing Values', ascending=False)

missing_values_df.head(6)
```
